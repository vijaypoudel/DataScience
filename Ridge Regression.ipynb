```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
import mglearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
```

```python
X,y = mglearn.datasets.load_extended_boston()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state =42)
ridge = Ridge().fit(X_train,y_train)
print("Training set score : {:.2f}".format(ridge.score(X_train,y_train)))
print("Test set score : {:.2f}".format(ridge.score(X_test,y_test)))
```

#### As seen the ridge gives a better score than linear regression. It lessen the chances of overfitting which was the case with linear regression.

## Role of Alpha in Ridge Regression

The ridge model makes a trade off between simplicity of the model (near zero coefficients) and its performance on the training set. 
Aplha is the one that decides how much the importance is place between the two. By default it is set to 1.0

```python
ridge10 = Ridge(5).fit(X_train,y_train)
print("Training set score : {:.2f}".format(ridge.score(X_train,y_train)))
print("Test set score : {:.2f}".format(ridge.score(X_test,y_test)))
```

```python
ridge01 = Ridge(0.1).fit(X_train,y_train)
print("Training set score : {:.2f}".format(ridge.score(X_train,y_train)))
print("Test set score : {:.2f}".format(ridge.score(X_test,y_test)))
```

#### Decreasing the alpha denotes decresing the restriction so it makes it equivalent to Linear regresson and hence we see a pattern similar to overfitting

```python
plt.plot(ridge.coef_,'ro',label="ridge alpha=1")
plt.plot(ridge10.coef_,'gs',label="ridge alpha=5")
plt.plot(ridge01.coef_,'bo',label="ridge alpha = 0.1")
plt.legend()
```
