### Population VS Sample 

Population is collection of all the items of interest denoted by "N" 
Sample is a subset of population which is denote by "n"

Population are hard to define and hard to observe
Sample are less time consuming and less complex

We always work on Sample data. Sample must be random and representative (acuractly define the item we are interested in)

### 3 Measures of Central Tendency

#### Mean  : 
Simple Average denoted by μ for population and x̄ for sample. Easily affected by outliers.

#### Median : 
To find median order the data and use (n+1)/2 to find the position which represent the median of the data. In case (n+1)/2 is not an integer the median is the simple average of (n/2 and (n+2)/2)th positioned datas.

#### Mode :
The reading that repeat itself the most in the given data set is said to be the mode.

### Measure of Asymmetry : Skewness
It represents whether the data is concentrated on one side. 
#### Positive Skewness / Right Skew : Mean > Median :
This means outliers are towards the right where the tail of the graph is .
 



#### Negative Skewness / Left Skew : Mean < Median :
This means outliers are towards the left where the tail of the graph is .
