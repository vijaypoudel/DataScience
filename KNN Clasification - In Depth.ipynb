```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
import mglearn
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
from sklearn.datasets  import load_breast_cancer
cancer = load_breast_cancer()
X_train, X_test, y_train ,y_test = train_test_split(cancer.data, cancer.target, random_state = 66)
```

```python
neighbours_selection = np.arange(1,11)
trainingAccuracy = []
testAccuracy = []
for neighbor in neighbours_selection:
    clf = KNeighborsClassifier(n_neighbors= neighbor)
    clf.fit(X_train,y_train)
    trainingAccuracy.append(clf.score(X_train,y_train))
    testAccuracy.append(clf.score(X_test,y_test))
plt.plot(neighbours_selection,trainingAccuracy,label="training accuracy")
plt.plot(neighbours_selection,testAccuracy,label="test accuracy")
plt.legend()
```

### Analyze the KNeighborsRegressor

```python
X,y = mglearn.datasets.make_wave(n_samples=40)
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state =0)
line =np.linspace(-3,3,10).reshape(-1,1)
```

```python
fig,axes = plt.subplots(1,3,figsize=(15,4))
for n_neighbors,ax in zip([1,3,9],axes):
    reg = KNeighborsRegressor(n_neighbors = n_neighbors)
    reg.fit(X_train,y_train)
    ax.plot(line,reg.predict(line))
    ax.set_xlabel("Feature")
    ax.set_ylabel("Target")
```
